SWAP cASE

```python
def swap_case(s):
    result = ''

    for char in s:
        if char.isupper():
            result += char.lower()
        elif char.islower():
            result += char.upper()
        else:
            result += char

    return result

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
def count_substring(string, sub_string):

    result = 0
    for i in range(len(string)):
        if sub_string == string[i : i + len(sub_string)]:
            # the substring taken from 'string' has the same lenght
            # of 'sub_string" and starts at the i-th character of
            # 'string'
            result += 1

    return result


if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()

    count = count_substring(string, sub_string)
    print(count)
```

Text Wrap

```python
import textwrap

def wrap(string, max_width):
    result = ''
    current_width = 0
    for char in string:
        result += char
        current_width += 1
        # when the current line has reached the max width
        # newline is inserted and the counter resetted
        if current_width == max_width:
            result += '\n'
            current_width = 0

    return result

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```

String Formatting

```python
def print_formatted(number):
    # your code goes here

    binary_size = len(str(bin(number))[2:])
    for i in range(1, number + 1):
        # the python native convertion is used then converted to
        # string and the first two characters representing that
        # the number is non-decimal (e.g. '0x') are cut

        binary = str(bin(i))[2:].rjust(binary_size, " ")
        decimal = str(i).rjust(binary_size, " ")
        octal = str(oct(i))[2:].rjust(binary_size, " ")
        hexadecimal = str(hex(i))[2:].upper().rjust(binary_size, " ")

        print(decimal + " " + octal + " " + hexadecimal + " " + binary)

if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

Alphabet Rangoli

```python
def print_rangoli(size):
    # your code goes here

    n = size * 2 - 1
    m = 2 * n -1

    # chr() converts a number to the representing character
    # in ASCII and 97 is the ASCII code for 'a', the other
    # letters follows alphabetically

    for i in range(- size + 1, size):
        j = abs(i)
        string = chr(97 + j)

        distance = size - j - 1
        for k in range(distance):
            letter = chr(97 + j + k + 1)
            string = letter + "-" + string + "-" + letter

        print(string.center(m, "-"))

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

The Minion Game

```python
def minion_game(string):
    # your code goes here

    vowels = ['A', 'E', 'I', 'O', 'U']
    stuart = 0
    kevin = 0

    for i in range(len(string)):
        substring_number = len(string) - i
        if string[i] in vowels:
            kevin += substring_number
        else:
            stuart += substring_number

    if kevin > stuart:
        print("Kevin", kevin)
    elif kevin == stuart:
        print("Draw")
    else:
        print("Stuart", stuart)

if __name__ == '__main__':
    s = input()
    minion_game(s)
```

Merge the Tools!

```python
def merge_the_tools(string, k):
    # your code goes here

    # separate into substrings
    substrings = []
    for i in range(len(string) // k):
        substrings.append(string[i * k : (i + 1) * k])

    for elem in substrings:
        result = []

        for char in elem:
            if char not in result:
                result.append(char)

        print(''.join(result))

if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```

Piling Up!

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

n_tests = int(input())

for i in range(n_tests):
    n = int(input())
    sizes = list(map(int, input().split(" ")))

    result = True
    last_removed = None
    for j in range(len(sizes)):
        if last_removed is not None and max(sizes[0], sizes[-1]) > last_removed:
            result = False
            break
        else:
            last_removed = max(sizes[0], sizes[-1])
            if last_removed == sizes[0]:
                sizes.pop(0)
            else:
                sizes.pop(-1)

    if result:
        print("Yes")
    else:
        print("No")
```

Time Delta

```python
#!/bin/python3

import math
import os
import random
import re
import sys

from datetime import datetime

# Complete the time_delta function below.
def time_delta(t1, t2):
    format_code = "%a %d %b %Y %H:%M:%S %z"
    timestamp1 = datetime.strptime(t1, format_code)
    timestamp2 = datetime.strptime(t2, format_code)

    result = abs(int((timestamp1 - timestamp2).total_seconds()))
    return str(result)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input())

    for t_itr in range(t):
        t1 = input()

        t2 = input()

        delta = time_delta(t1, t2)

        fptr.write(delta + '\n')

    fptr.close()
```

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

from html.parser import HTMLParser

class Parser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print("Start :", tag)
        for elem in attrs:
            print("-> " + str(elem[0]) + " > " + str(elem[1]))
    def handle_endtag(self, tag):
        print("End   :", tag)
    def handle_startendtag(self, tag, attrs):
        print("Empty :", tag)
        for elem in attrs:
            print("-> " + str(elem[0]) + " > " + str(elem[1]))

parser = Parser()

n = int(input())

for i in range(n):
    string = input()
    parser.feed(string)
```

HTML Parser - Part 2

```python
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_comment(self, data):
        comments = data.split("\n")
        if len(comments) == 1:
            print(">>> Single-line Comment")
        else:
            print(">>> Multi-line Comment")

        for elem in comments:
            print(elem)

    def handle_data(self, data):
        if data != "\n":
            print(">>> Data")
            print(data)

html = ""
for i in range(int(input())):
    html += input().rstrip()
    html += '\n'

parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

Matrix Script

```python
#!/bin/python3

import math
import os
import random
import re
import sys

first_multiple_input = input().rstrip().split()

n = int(first_multiple_input[0])
m = int(first_multiple_input[1])

matrix = []

for _ in range(n):
    matrix_item = input()
    matrix.append(matrix_item)

# extract codified word
string = ""
for i in range(m):
    for j in range(n):
        string += matrix[j][i]

# extract starting symbols
start_string = ""
start_index = 0
while start_index < len(string) and not string[start_index].isalpha():
    start_string += string[start_index]
    start_index += 1

# extract ending symbols
end_string = ""
end_index = len(string) - 1
while end_index > start_index and not string[end_index].isalpha():
    end_string = string[end_index] + end_string
    end_index -= 1

# extract words
pattern = r"(?P<word>\w+)[$#% \s]*"
wordlist = re.findall(pattern, string[start_index : end_index + 1])
result = ""

for elem in wordlist:
    result += elem + " "

result = result[0 : len(result) - 1]

# concatenate and print result
result = start_string + result + end_string
print(result)
```

XML2 - Find the Maximum Depth

```python
import xml.etree.ElementTree as etree

maxdepth = 0
def depth(node, level):
    global maxdepth
    level += 1

    if level > maxdepth:
        maxdepth = level

    for elem in node:
        depth(elem, level)

if __name__ == '__main__':
    n = int(input())
    xml = ""
    for i in range(n):
        xml =  xml + input() + "\n"
    tree = etree.ElementTree(etree.fromstring(xml))
    depth(tree.getroot(), -1)
    print(maxdepth)
```

```python
def wrapper(f):
    def fun(l):
        for i in range(len(l)):
            if len(l[i]) == 13:
                l[i] = l[i][3 : 13]
            elif len(l[i]) == 12:
                l[i] = l[i][2 : 12]
            elif len(l[i]) == 11:
                l[i] = l[i][1 : 11]
            else:
                l[i] = l[i][0 : 10]

        result = []
        for elem in l:
            string = "+91 " + elem[0 : 5] + " " + elem[5 : 11]
            result.append(string)

        f(result)

    return fun

@wrapper
def sort_phone(l):
    print(*sorted(l), sep='\n')

if __name__ == '__main__':
    l = [input() for _ in range(int(input()))]
    sort_phone(l)
```

Number Line Jumps

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'kangaroo' function below.
#
# The function is expected to return a STRING.
# The function accepts following parameters:
#  1. INTEGER x1
#  2. INTEGER v1
#  3. INTEGER x2
#  4. INTEGER v2
#

def kangaroo(x1, v1, x2, v2):
    if x1 == x2:
        return "YES"
    elif v1 == v2:
        # if relative speed is null they never meet
        return "NO"
    elif (x1 > x2 and v1 > v2) or (x2 > x1 and v2 > v1):
        # if the further is faster then it's impossible
        return "NO"

    relative_speed = v1 - v2
    relative_distance = x1 - x2
    if relative_distance % relative_speed == 0:
        return "YES"
    else:
        return "NO"




if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    first_multiple_input = input().rstrip().split()

    x1 = int(first_multiple_input[0])

    v1 = int(first_multiple_input[1])

    x2 = int(first_multiple_input[2])

    v2 = int(first_multiple_input[3])

    result = kangaroo(x1, v1, x2, v2)

    fptr.write(result + '\n')

    fptr.close()
```

Recursive Digit Sum

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'superDigit' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. STRING n
#  2. INTEGER k
#

def superDigit(n, k):
    value = 0
    for char in n:
        value += int(char)
    value *= k

    while len(str(value)) > 1:
        new_value = 0
        for char in str(value):
            new_value += int(char)
        value = new_value

    return value

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    first_multiple_input = input().rstrip().split()

    n = first_multiple_input[0]

    k = int(first_multiple_input[1])

    result = superDigit(n, k)

    fptr.write(str(result) + '\n')

    fptr.close()
```
